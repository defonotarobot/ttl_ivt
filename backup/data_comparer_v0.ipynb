```python
colnum = df1.shape[1]

def compare(merged_table):
    
    df1 = merged_table.drop('id', axis = 1)
    result = merged_table[['id']]
    col = 2 #start from third col
#     n = 1
    for i in range(0,colnum-1):
        com_col = df1.iloc[:,col-2:col] #creating a comparison df with 2 cols 
        com_col['COM_' + com_col.columns[0][:-2]] = com_col.iloc[:,0] == com_col.iloc[:,1] #
        col += 2
        result = pd.concat([result, com_col], axis = 1)
#         n+=3
                   
    return result
```

```python
def prefix(tbl):
    result = tbl
    for col in result.columns:
        new_col = ''
        if '_x' in col:
            new_col = env1 + '_' + col.removesuffix('_x')
        elif '_y' in col:
            new_col = env2 + '_' + col.removesuffix('_y')
        else:
            new_col = col
#         print(new_col)
        result = result.rename(columns = {col:new_col})
                   
    return result
```
